```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
x = np.arange(0,np.pi*2,0.1)
x  # independent 
```

```python
y = np.sin(x)
y # dependent
```

```python
# if both x and y is dependent then there can be change in grap
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine of Angle')
plt.title('Sine Function')
plt.show()
```

```python
z = np.cos(x)
```

```python
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('cos of Angle')
plt.title('Cos Function')
```

```python
v = np.tan(x)
```

```python
plt.plot(x,v)
plt.xlabel('angle')
plt.ylabel('Tan of Angle')
plt.title('Tan Function')
```

```python
fig = plt.figure() # by default value of fig size
```

```python
fig = plt.figure()
a = fig.add_axes([0,0,.5,.5])   # Figure size not for graph x and y values
a.plot(x,y,linewidth = 6)     # linewidth default value is 1
a.set_xlabel('Angle')
a.set_ylabel('Sine of Angles')
a.set_title('Sine Function')
```

# Sub Plots

```python
fig , ax = plt.subplots(2,3)
```

```python
fig , ax = plt.subplots(2,2)
plt.tight_layout() # this is to avoid overlapping
ax[0][0].plot(x,y)
ax[0][1].plot(x,z)
ax[1][0].plot(y,z)
ax[1][1].plot(x,v)
ax[0][0].set_title('Sine Function')
ax[0][1].set_title('Cosine Function')
ax[1][0].set_title('Sine Cosine Function')
ax[1][1].set_title('Tan Function')
```

```python
plt.pie(y,labels = x)
plt.legend(loc=3)
plt.show()
```

```python
li = [0.2,0.1,0,0,0]
plt.pie(y,labels = x,explode = li)
plt.show()
```

```python
l =[]
for i in range(len(y)):
    if y[i]==min(y):
        l.append(0.1)
    elif y[i]==max(y):
        l.append(0.2)
    else:
        l.append(0)
l
```

```python
l1 = [0.1 if i == min(y) else 0.2 if i == max(y) else 0 for i in y]

plt.pie(y,labels = x,explode = l1 , autopct ='%.2f%%')
plt.show()
```
