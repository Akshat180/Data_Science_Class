### The InterQuartile Range
- IQR = Q3 - Q1
- MaxIQR = Q3 + 1.5*IQR
- MinIQR = Q1 + 1.5*IQR
- Outliers are those jo max iqr se jada or min iqr se kum ho 
- An outlier is a single data point that goes far outside the average value of a group of statistics.

### Five Number Summary
- Min
- Q1
- Q2
- Q3
- Max

Min, Max and Q2[median] will help to find out skeawness and using Q3 and Q1 we will find where oulier lies
