```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
x = np.arange(0,np.pi*2,0.1)
x  # independent 
```

```python
y = np.sin(x)
y # dependent
```

```python
# if both x and y is dependent then there can be change in grap
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine of Angle')
plt.title('Sine Function')
plt.show()
```

```python
z = np.cos(x)
```

```python
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('cos of Angle')
plt.title('Cos Function')
```

```python
v = np.tan(x)
```

```python
plt.plot(x,v)
plt.xlabel('angle')
plt.ylabel('Tan of Angle')
plt.title('Tan Function')
```

```python
fig = plt.figure() # by default value of fig size
```

```python
fig = plt.figure()
a = fig.add_axes([0,0,.5,.5])   # Figure size not for graph x and y values
a.plot(x,y)
a.set_xlabel('Angle')
a.set_ylabel('Sine of Angles')
a.set_title('Sine Function')
```

```python
x
y
z
fig = plt.figure()
axis = fig.add_axes([0,0,.5,.5])

axis.plot(x,y, color = 'lightgreen',linestyle = '--')
axis.plot(x,z,color = 'lightblue', marker = '*')
```

```python
fig = plt.figure()
axis = fig.add_axes([0,0,.8,.8])

axis.plot(x,y, color = 'green')
axis.plot(x,v,color = 'blue')
```

```python
fig = plt.figure()
axis = fig.add_axes([0,0,.8,.8])

axis.plot(x,y, color = 'green', linestyle = '-.') # use color name or hex code ex #e35da4
axis.plot(x,v,color = 'lightblue', linestyle = '--')
axis.plot(x,z,color = '#e35da4', marker ='*')
```

```python
x
y
z
fig = plt.figure()
axis = fig.add_axes([0,0,.8,.8])

axis.plot(x,y, color = 'lightgreen',linestyle = '--', label = 'sine')
axis.plot(x,z,color = 'blue', marker = '*', label = 'cosine')
axis.legend(loc = 1) # or we can use Loc = 'upper right'
# axis.plot(x,x)
```

```python
x
y
z
fig = plt.figure()
axis = fig.add_axes([0,0,.8,.8])

axis.plot(x,y, color = 'lightgreen',linestyle = '--')
axis.plot(x,z,color = 'blue', marker = '*')
axis.legend(labels = ['Sine','Cosine'], loc = 1) # or we can use Loc = 'upper right'
# axis.plot(x,x)    #['1st graph','2nd graph']
```
