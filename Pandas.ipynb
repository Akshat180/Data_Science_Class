## Combining data in Pandas

```python
a = [2,5,8,1,9,3,8,0,4,6]
b = [4,7,0,2,4,1,9,-1]
x = pd.Series(a)
y = pd.Series(b)                    # comparing both list and printing the greater one 

z = x.combine(y,(lambda a1, b1: a1 if a1>b1 else b1))
list(z)
```

```python
z = x.combine(y,(lambda a1, b1: a1 + b1 ))   #  adding both the list 
list(z)
```

## Merge in Pandas

```python
d1 =  {'id':[1,2,3,4,5],
       'Name': list('ABCDE'),
       'Subject':['sub1','sub2','sub4','sub6','sub5']}
d2 =  {'id':[1,2,3,4,5],
       'Name': list('FGHIJ'),
       'Subject':['sub2','sub4','sub3','sub6','sub5']}
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,  on = 'id')      # _x and _y are default suffix
```

```python
pd.merge(df1,df2,  on = 'Name')   # no value due both values in name are exlusive
#                                    agar how ni de toh by defaut voh inner lega
```

```python
pd.merge(df1,df2,  on = 'Subject', how = 'inner')  # how = in which way we want inner = intersection , outer = union
                                                    # inner = dono mai hai , outer = dono mai ni hai 
```

```python
pd.merge(df1,df2,  on = 'Subject', how = 'outer')  
```

```python
pd.merge(df1,df2,  on = 'Name', how = 'outer') 
```

```python
pd.merge(df1,df2,  on = 'Subject', how = 'left')  # left = left mai jo DF hai ex: df1 voh pura print hoga 
#                                                          or df2 mai jo common hoga voh print hoga
#                                                   right = isme ulta hoga
```

```python
pd.merge(df1,df2,  on = ['id','Subject'], how = 'inner')  
```
